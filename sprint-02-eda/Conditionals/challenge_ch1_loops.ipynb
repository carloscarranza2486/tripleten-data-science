### Contexto del problema:
Eres analista de datos de una tienda online y necesitas procesar las ventas de la semana pasada. Tienes datos de diferentes productos y debes automatizar el análisis usando bucles for.

Datos proporcionados:
Copia y pega este código en tu primera celda para crear los datasets:

```python
# Datos de ventas diarias (en unidades vendidas)
productos = ['Laptop', 'Mouse', 'Teclado', 'Monitor', 'Auriculares']
ventas_lunes = [5, 23, 15, 8, 12]
ventas_martes = [7, 19, 18, 6, 15]
ventas_miercoles = [3, 25, 12, 10, 9]
ventas_jueves = [8, 21, 20, 7, 18]
ventas_viernes = [12, 30, 25, 15, 22]

# Precios por unidad (en dólares)
precios = [899, 25, 45, 299, 79]

# Lista de días para referencia
dias = ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes']
```

### Tarea 1: Ventas totales por producto

Usa un bucle for para calcular cuántas unidades se vendieron de cada producto durante toda la semana. Almacena los resultados en una nueva lista llamada ventas_totales.

```python
# 1. Creamos la lista vacía donde guardaremos los 5 totales
ventas_totales = []

# 2. Iniciamos el ciclo. 
# len(productos) es 5. range(5) nos dará: 0, 1, 2, 3, 4
for i in range(len(productos)):
    
    # 3. Sumamos lo que hay en la posición 'i' de cada día
    # Si i vale 0, estamos sumando todo lo de la Laptop
    suma_semanal = (ventas_lunes[i] + 
                    ventas_martes[i] + 
                    ventas_miercoles[i] + 
                    ventas_jueves[i] + 
                    ventas_viernes[i])
    
    # 4. Agregamos esa suma a nuestra lista de resultados
    ventas_totales.append(suma_semanal)

# Imprimimos para verificar
print("Ventas totales por producto:", ventas_totales)

for i in range(len(productos)):
    
    nombre = productos[i]         # Saca "Laptop" (Posición 0)
    total = ventas_totales[i]     # Saca 35 (Posición 0)
    
    print(f"El producto {nombre} vendió {total} unidades")

```

### Tarea 2: Ingresos por producto

Calcula cuánto dinero generó cada producto multiplicando las ventas totales por su precio. Usa un bucle for y almacena los resultados en una lista llamada ingresos_productos.

```python
ingresos_productos = []

for i in range(len(productos)):
    dinero_productos = ventas_totales[i] * precios[i]
    ingresos_productos.append(dinero_productos)
    print(f'El {productos[i]} generó {dinero_productos} dólares')
```

### Tarea 3: Reporte diario

Para cada día de la semana, calcula:

El total de unidades vendidas ese día (suma de todos los productos)
El producto más vendido ese día
El ingreso total de ese día
Usa bucles for y muestra los resultados con print().

```python
todas_las_ventas = [ventas_lunes, ventas_martes, ventas_miercoles, ventas_jueves, ventas_viernes]

for i in range(len(dias)):
    dia = dias[i] # obtenemos el día exacto en cada iteraqción
    lista_dia = todas_las_ventas[i] # extraemos la lista de la matriz para tener solo los datos de unidades vendidas de este día
    unidades_vendidas = sum(lista_dia) # calculamos unidades vendidas
    mas_vendido = max(lista_dia) # definimos la mayor cantidad de unidades ese día
    indice_mas_vendido = lista_dia.index(mas_vendido) # encontramos el índice del producto en la lista del día
    producto_mas_vendido = productos[indice_mas_vendido] # encontramos el nombre del producto en la lista del día

    ingreso_dia = 0 # definimos el punto de partida del día en término de ingresos

    for j in range(len(productos)):
        cantidad = lista_dia[j] # definimos cantidad de unidades a partir de cantidad de indexación
        precio = precios[j] # extraemos precio de producto según iteración
        ingreso_dia += cantidad*precio # calculamos ingreso total del día multiplicando unidades por precio de cada producto
        

    print(f'\n-------------Reporte del día {dia}-------------') # imprimimos reporte del día
    print(f'El total de unidades vendidas del día {dia} son {unidades_vendidas}')
    print(f'El producto más vendido fue: {producto_mas_vendido}')
    print(f'Los ingresos totales del día son {ingreso_dia}')
```
